#### Treating Outliers

```python
for columns in df:
    df.boxplot([columns])
    plt.show()
```

```python
#INTERQUANTILE RANGE [IQR]

def outlier_treatment(col):
    sorted(col)
    Q1,Q3=np.percentile(col, [25,75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
lower_range,upper_range=outlier_treatment(df['Pregnancies'])
print('Lower Range:',lower_range)
print('Upper Range:',upper_range)
```

```python
df['Pregnancies']=np.where(df['Pregnancies']>upper_range,upper_range,np.where(df['Pregnancies']<lower_range,lower_range,df['Pregnancies']))
```
